### Import Data

```python
# Import our libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import from sklearn.
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('updated df 05-20-2020.csv')
```

```python
df.fillna(0, inplace=True)
```

```python
X = df.drop(columns=['Player', 'Pos', 'Team', 'Year', 'Age'])
```

### Standardize our Data

```python
ss = StandardScaler()
```

```python
X_scaled = ss.fit_transform(X)
```

```python
pca = PCA(n_components=5, random_state=42)
```

```python
pca.fit(X)
```

```python
Z_features = pca.transform(X)
```

### Try K-Means with less clusters

```python
km = KMeans(n_clusters=7, random_state=42)
km.fit(Z_features)
```

```python
df['Cluster Labels'] = km.labels_
```

```python
df['Cluster Labels'].value_counts()
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(Z_features[:, 0], Z_features[:, 1],
            c=km.labels_, edgecolor='none', alpha=0.8,
            cmap=plt.cm.get_cmap('nipy_spectral'))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
df.groupby('Cluster Labels').mean()
```
