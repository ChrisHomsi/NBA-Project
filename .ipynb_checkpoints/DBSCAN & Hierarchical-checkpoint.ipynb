```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('updated df 05-20-2020.csv')
```

```python
df.fillna(0, inplace=True)
```

```python
X = df.drop(columns=['Player', 'Pos', 'Team', 'Year', 'Age'])
```

```python
ss = StandardScaler()
```

```python
X_scaled = ss.fit_transform(X)
```

```python
db_model = DBSCAN(eps=4.1, min_samples=4)
db_model.fit(X_scaled)
```

### Only creating noise cluster?

```python
set(db_model.labels_)
```

```python
silhouette_score(X_scaled, db_model.labels_)
```

```python
df['cluster'] = db_model.labels_
```

```python
df.head(3)
```

```python
df['cluster'].value_counts(normalize=True)
```

```python
df.groupby('cluster').mean()
```

### DBSCAN is not great, Hierarchical Clustering Next! (Agglomerative Clustering)

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=8, affinity='euclidean', linkage='ward')  
model_ac = ac.fit(X_scaled)
```

```python
model_ac.labels_
```

```python
df['cluster'] = model_ac.labels_
```

```python
df['cluster'].value_counts()
```

```python
df.groupby('cluster').mean()
```

### Label Each cluster accordingly

```python
labels = {
    0 : 'Bench Player',
    1 : 'Defensive Big Men',
    2 : '2-way Wing', 
    3 : 'Secondary floor general',
    4 : 'Elite Overall Center',
    5 : 'Shooting Specialist',
    6 : 'Defensive Specialist',
    7 : 'Elite Overall Guard'
}
```

```python
df['cluster'] = df['cluster'].map(labels)
```
