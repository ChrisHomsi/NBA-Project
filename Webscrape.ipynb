```python
years_to_scrape = ['2020','2019', '2018', '2017', '2016']

'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'.format(years_to_scrape[0])
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
list_of_soups = []

for year in years_to_scrape:
    url = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'.format(years_to_scrape)
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    list_of_soups.append(soup)
```

```python
list_of_soups[0].find('tr')
```

```python
stats = []

for soup in (list_of_soups):
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    stats.append(player_stats)
```

```python
list_of_soups[0].find_all('tr')
```

```python
df = pd.DataFrame(player_stats, columns= headers)
df.shape
```
